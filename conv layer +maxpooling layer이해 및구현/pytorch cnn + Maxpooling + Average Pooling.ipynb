```python
# Correlation Calculation
import numpy as np
N, n_H, n_W, n_C = 1, 28, 28, 1
n_filter =1
k_size = 3


image = torch.randn(size=(N, n_C, n_H, n_W))

conv = nn.Conv2d(in_channels=n_filter, out_channels=1, kernel_size=k_size)
y = conv(image)
weight, bias = conv.weight, conv.bias
print('='*100)

image = image.detach().numpy().squeeze()
weight = weight.squeeze()
# print(image.size())
# print(weight.size())
# print(bias.size())


y_man = np.zeros(shape=(n_H - k_size + 1, n_W - k_size + 1))


for i in range(n_H - k_size + 1):
  for j in range(n_W - k_size + 1):
    window = image[i: i+k_size, j: j+k_size]
    y_man[i, j] = np.sum(window * weight) + bias


y_man = np.zeros(shape=(n_H - k_size + 1, n_W - k_size + 1))

# Convolution 3D 
for i in range(n_H - k_size + 1):
  for j in range(n_W - k_size + 1):
    window = image[i: i+k_size, j: j+k_size, :]
    y_man[i, j] = np.sum(window * weight) + bias

```

```python
n, nh,nw,nc = 1, 5, 5, 3 
n_filter = 3
k_size= 4 


# batch_size, channel, height, width 
images = torch.randn(size=(n, nc, nh,nw))
convv = nn.Conv2d(in_channels=nc, out_channels=n_filter, kernel_size=k_size)
y = convv(images)

c_weight, c_bias = convv.weight, convv.bias

# 5 x 5인 RGB 값을 가진 이미지가 Convolution에 들어갔다. 그래서 conv의 weight와 bias를 해석해보자면, 4 x 4 window와 3개의 필터가 각각 연산이 되어 총 3개의 이미지가 뽑히는것이다.
# 그래서 y의 shape를 보면 한장의 이미지가 나오는데 1장의 이미지의 rgb값 3개 연산되어 나왔고 2x2사이즈로 줄어들었다. 
print('images shape: ', images.size())
print('weight shape: ', c_weight.size())
print('bias shape: ', c_bias.size())
print('y shape: ', y.size())

yy = y.squeeze()
print(yy.shape)
Y_man = np.zeros(shape=(nh - k_size + 1, nw - k_size + 1, n_filter))
print(Y_man.shape)


images = images.squeeze()
images = np.transpose(images, (2,1, 0))
print(images.shape)
for c in range(n_filter):
  c_W = c_weight[:, c, :]
  c_W = c_W.detach().numpy()
  c_W = np.transpose(c_W, (2, 1, 0))
  print(c_W.shape)
  c_B = c_bias[c]

  for h in range(nh - k_size + 1):
    for j in range(nw - k_size + 1):
      window = images[h:h+k_size, j:j+k_size, :]
      window = window.detach().numpy()
      con_plus = np.sum(c_W*window) + c_B
      # print('Weight shape: ', c_W.shape, 'window shape: ', window.shape)
      Y_man[h,j, c] =  con_plus
```

```python
#model with Conv Layers 

n_neurons = [10,20, 30]


model = nn.Sequential(
    nn.Conv2d(in_channels=3, out_channels=n_neurons[0], kernel_size=3),
    nn.ReLU(),
    nn.Conv2d(in_channels=n_neurons[0], out_channels=n_neurons[1], kernel_size=3),
    nn.ReLU(),
    nn.Conv2d(in_channels=n_neurons[1], out_channels=n_neurons[2], kernel_size=3)
)

x = torch.randn(size = (32, 3, 28, 28))

y = model(x)
print(f'input size: {x.size()}')
print(f'output size: {y.size()}')
```

```python
class Testmodel(nn.Module):
  def __init__(self):
    super(Testmodel, self).__init__()


    n_neurons = [10,20, 30]
    self.conv = nn.Sequential(
        nn.Conv2d(in_channels=3, out_channels=n_neurons[0], kernel_size=3),
        nn.ReLU(),
        nn.Conv2d(in_channels=n_neurons[0], out_channels=n_neurons[1], kernel_size=3),
        nn.ReLU(),
        nn.Conv2d(in_channels=n_neurons[1], out_channels=n_neurons[2], kernel_size=3)    
    )


  def forward(self, x):
    print(self.conv[0])
    conv1_weight, conv1_bias = self.conv[0].weight, self.conv[0].bias
    print('Conv1 layer shape: ',conv1_weight.size(), conv1_bias.size())
    print()
    print('Conv2 layer shape:', self.conv[2])
    conv2_weight, conv2_bias = self.conv[2].weight, self.conv[2].bias
    print('Conv1 layer shape: ',conv2_weight.size(), conv2_bias.size())
    print()

    print('Conv2 layer shape:', self.conv[4])
    conv3_weight, conv3_bias = self.conv[4].weight, self.conv[4].bias
    print('Conv1 layer shape: ',conv3_weight.size(), conv3_bias.size())
    output = self.conv(x)

    return output 

```

```python
# Max Pooling 
import numpy as np

L, f, s  = 10, 2, 1 


x = torch.randn(size=(1, 1, 10))
print(x.size())
pooled_max = nn.MaxPool1d(kernel_size=f, stride=s)
pooledd_max =pooled_max(x)


print('x: {}\n{}'.format(x.size(), x.numpy().flatten()))

print('pooled_max(Pytorch): {}\n{}'.format(pooledd_max.size(), pooledd_max.numpy().flatten()))




## window을 한 이미지에서 돌아다니면서 max값을 가져오는 방법
xx = x.numpy().flatten()

print()
poool_man = np.zeros(shape=(L - f + 1))

for i in range(L - f + 1):
  windows = xx[i:i+f]
  print(windows)
  poool_man[i] = np.max(windows)


print(f'poool_man shape: {poool_man.shape}, poool_man: {poool_man}')
```

```python

# Average Pooling 
L, f, s = 10, 2, 1

# batchsize, channel, datas
x1 = torch.randn(size=(1, 1, L))

pool_avg = nn.AvgPool1d(kernel_size = f, stride=s)
pooled_avg = pool_avg(x1)

print(f'x shape: {x1.size()}\nx: {x1}')
print(f'Average Poold_1D: {pooled_avg}\nAverage Poold1D shape:{pooled_avg.size()}')



x1 = x1.numpy().flatten()


avg_pool = np.zeros(shape=(L-f + 1))


for channel in range(L-f+1):
  wind = x1[channel: channel+f]

  avg_pool[channel] = np.mean(wind)
print(f'numpy avg pool: {avg_pool} numpy shape: {avg_pool.shape}')
```

```python

# Max Pooling 2D 

N, n_H, n_W, C = 1, 5, 5, 1
f, s = 2, 1

x2 = torch.randn(size=(N, C, n_H, n_W))
print(x2.size())



pool_max = nn.MaxPool2d(kernel_size=2, stride=1)
pooled_max = pool_max(x2)


print(f'x2 shape: {x2.size()}, x2: {x2}')
print(f'maxpool 2D: {pooled_max}, maxpool 2D shape:{pooled_max.shape}')
```

```python
# Max Pooling row - level 
x2_sqze = x2.detach().numpy().squeeze()
print(x2_sqze.shape)


pools2d = np.zeros(shape=(n_H - f + 1, n_W -f + 1))


for i in range(n_H - f +1):
  for j in range(n_W -f +1):

    windows = x2_sqze[i:i+f, j:j+f]
    maxss = np.max(windows)
    pools2d[i, j] = maxss


print(f'numpy get 2D : {pools2d}, 2D shape: {pools2d.shape}')
```
