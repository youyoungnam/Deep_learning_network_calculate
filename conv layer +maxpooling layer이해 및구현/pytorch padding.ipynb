```python
import torch 
import torch.nn as nn
import numpy as np 
import math 
```

```python
# Maxpooling 3D
N, n_H, n_W, n_C = 1, 5, 5, 3
f, s = 2,2 




x = torch.randn(size=(N, n_C, n_H, n_W))
print(x.size())

# window가 지나가면서 max pool가 되고 각 채널별로 뽑히게 된다. 그래서 채널은 그대로 height와 width은 변한다. 
pool_max = nn.MaxPool2d(kernel_size=f, stride=s)
pooled_max = pool_max(x)
print(pooled_max.size())
print(pooled_max)


xx = x.detach().numpy().squeeze()
xx = np.transpose(xx, (2, 1, 0))

print(xx.shape)

n_H_ = math.floor((n_H - f)/s+1)
n_W_ = math.floor((n_W - f)/s+1)
print(n_H_, n_W_)


pooled_max_man = np.zeros(shape= (n_H_, n_W_, n_C))
print(pooled_max_man)


for c in range(n_C):
  c_images = xx[:, :, c]

  h_ = 0
  for h in range(0,n_H -f + 1, s):
    w_ = 0 
    for w in range(0, n_W -f + 1, s):
      window = c_images[h:h+f, w:w+f]
      pooled_max_man[h_,w_, c] = np.max(window)

      w_ +=1
    h_ +=1
print(pooled_max_man.shape)
print(pooled_max_man)
```

```python
zero_images = torch.randn(size=(1, 3, 3, 3))
print(zero_images.size())


zero_padding = nn.ZeroPad2d(padding=2)
zeroed_padding = zero_padding(zero_images)
print(zeroed_padding)
```

```python
# conv 2d + padding
images = torch.randn(size=(1, 3, 28, 28))

conv = nn.Conv2d(in_channels=3, out_channels=1, kernel_size=3, padding='same')
y = conv(images)
print(images.size())
print(y.size())
```

```python
# stride + conv2d


images = torch.randn(size=(1, 3, 28, 28))
print('image size: ', images.size())

convs = nn.Conv2d(in_channels=3, out_channels=1, kernel_size= 3 , padding='valid', stride=2)
cv = convs(images)
print('output size: ', cv.size())
```

```python
# stride + pooling 
images = torch.randn(size=(1, 3, 28, 28))
print('image size: ', images.size())

pool = nn.MaxPool2d(kernel_size=3, stride=2)
cv = pool(images)
print('output size: ', cv.size())
```
